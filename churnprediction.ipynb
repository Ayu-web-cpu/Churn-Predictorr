```python
def predict_churn_from_user_():


    # This is the order in which model expects the features
    feature_order = [
        'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
        'PhoneService', 'Contract', 'PaperlessBilling', 'MonthlyCharges',
        'TotalCharges', 'MultipleLines_No phone service', 'MultipleLines_Yes',
        'OnlineSecurity_No internet service', 'OnlineSecurity_Yes',
        'OnlineBackup_No internet service', 'OnlineBackup_Yes',
        'DeviceProtection_No internet service', 'DeviceProtection_Yes',
        'TechSupport_No internet service', 'TechSupport_Yes',
        'StreamingTV_No internet service', 'StreamingTV_Yes',
        'StreamingMovies_No internet service', 'StreamingMovies_Yes',
        'InternetService_Fiber optic', 'InternetService_No',
        'PaymentMethod_Credit card (automatic)',
        'PaymentMethod_Electronic check', 'PaymentMethod_Mailed check'
    ]

    user_input = {}

    print("\n--- Please enter the customer details step by step ---\n")

    # Binary features with 0 or 1 input
    print("Enter 0 for 'No' and 1 for 'Yes' where applicable.\n")
    binary_questions = {
        'gender': "Gender: Enter 0 for Female or 1 for Male: ",
        'SeniorCitizen': "Is the customer a senior citizen? Enter 0 for No or 1 for Yes: ",
        'Partner': "Does the customer have a partner? Enter 0 for No or 1 for Yes: ",
        'Dependents': "Does the customer have dependents? Enter 0 for No or 1 for Yes: ",
        'PhoneService': "Does the customer have phone service? Enter 0 for No or 1 for Yes: ",
        'PaperlessBilling': "Is the billing paperless? Enter 0 for No or 1 for Yes: "
    }

    for key, question in binary_questions.items():
        while True:
            try:
                answer = int(input(question))
                if answer == 0 or answer == 1:
                    user_input[key] = answer
                    break
                else:
                    print("Please enter only 0 or 1.")
            except ValueError:
                print("Invalid input. Please enter a number (0 or 1).")

    # Contract type input
    print("\nChoose the contract type:")
    print("0 - Month-to-month\n1 - One year\n2 - Two year")
    while True:
        try:
            contract = int(input("Enter contract type number: "))
            if contract in [0, 1, 2]:
                user_input['Contract'] = contract
                break
            else:
                print("Only enter 0, 1 or 2.")
        except ValueError:
            print("Enter a valid number (0, 1, or 2).")

    # Numeric inputs
    numeric_features = {
        'tenure': "Enter number of months the customer has stayed: ",
        'MonthlyCharges': "Enter the monthly charges: ",
        'TotalCharges': "Enter the total charges: "
    }

    for key, question in numeric_features.items():
        while True:
            try:
                value = float(input(question))
                user_input[key] = value
                break
            except ValueError:
                print("Please enter a valid number.")

    # One-hot encoded features (Yes or No)
    print("\nNow answer the following features that are already one-hot encoded. Enter 1 for Yes, 0 for No.\n")
    ohe_features = [
        'MultipleLines_No phone service', 'MultipleLines_Yes',
        'OnlineSecurity_No internet service', 'OnlineSecurity_Yes',
        'OnlineBackup_No internet service', 'OnlineBackup_Yes',
        'DeviceProtection_No internet service', 'DeviceProtection_Yes',
        'TechSupport_No internet service', 'TechSupport_Yes',
        'StreamingTV_No internet service', 'StreamingTV_Yes',
        'StreamingMovies_No internet service', 'StreamingMovies_Yes',
        'InternetService_Fiber optic', 'InternetService_No',
        'PaymentMethod_Credit card (automatic)',
        'PaymentMethod_Electronic check', 'PaymentMethod_Mailed check'
    ]

    for key in ohe_features:
        while True:
            try:
                answer = int(input(f"{key.replace('_', ' ')} (0=No, 1=Yes): "))
                if answer in [0, 1]:
                    user_input[key] = answer
                    break
                else:
                    print("Only enter 0 or 1.")
            except ValueError:
                print("Please enter 0 or 1.")

    # Prepare the data in the format model expects
    input_df = pd.DataFrame([[user_input[feature] for feature in feature_order]], columns=feature_order)

    # Apply scaling (if used)
    input_scaled = scaler.transform(input_df)

    # Predict churn using trained model
    prediction = lr.predict(input_scaled)[0]

    print("\nPrediction Result:")
    if prediction == 1:
        print("The customer is likely to CHURN.")
    else:
        print("The customer is NOT likely to churn.")
```
